```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('C:/Users/Omar/OneDrive/Vs Code/ML_Proj/loan_eligibility_project/data/loan-train.csv')
print(f"Shape: {df.shape}")
df.head()
```

```python
df.drop('Loan_ID', axis=1, inplace=True)
```

```python
gender_mode = df['Gender'].mode()[0]
df['Gender'].fillna(gender_mode, inplace=True)

married_mode = df['Married'].mode()[0]
df['Married'].fillna(married_mode, inplace=True)

dependents_mode = df['Dependents'].mode()[0]
df['Dependents'].fillna(dependents_mode, inplace=True)

df['Self_Employed'].fillna('No', inplace=True)
```

```python
loanamount_median = df['LoanAmount'].median()
df['LoanAmount'].fillna(loanamount_median, inplace=True)

loan_term_mode = df['Loan_Amount_Term'].mode()[0]
df['Loan_Amount_Term'].fillna(loan_term_mode, inplace=True)

credit_history_mode = df['Credit_History'].mode()[0]
df['Credit_History'].fillna(credit_history_mode, inplace=True)
```

```python
binary_features = ['Gender', 'Married', 'Education', 'Self_Employed']

for feature in binary_features:
    df[feature] = df[feature].map({'Yes': 1, 'No': 0, 'Male': 1, 'Female': 0, 'Graduate': 1, 'Not Graduate': 0})

multi_class_features = ['Dependents', 'Property_Area']

df = pd.get_dummies(df, columns=multi_class_features, drop_first=True)

print(f"Shape: {df.shape}")
df.head()
```

```python
skewed_features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']

for feature in skewed_features:
    df[feature] = np.log1p(df[feature])
print(f"Shape: {df.shape}")
df.head()
```

```python
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status'].map({'Y': 1, 'N': 0}) 

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X.to_csv('../processed_data/X_processed.csv', index=False)
y.to_csv('../processed_data/y_processed.csv', index=False)

print("✅ Processed data saved successfully!")
```
